```python
from sklearn.datasets import make_multilabel_classification
from sklearn.model_selection import train_test_split
from sklearn.multioutput import MultiOutputClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
```

```python
df=pd.read_csv("final_update.csv")
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
col = ['prognosis','Ayurvedic Category']
for i in col:
    df[i] = le.fit_transform(df[i]).astype(int)
```

```python
df
```

```python
# Assuming you have already fitted and transformed your data using LabelEncoder
# le = LabelEncoder()
# df['prognosis'] = le.fit_transform(df['prognosis']).astype(int)

# Get the mapping of encoded labels to original string values
label_mapping = dict(zip(range(len(le.classes_)), le.classes_))

# Print the mapping
print("Encoded Label : Original String Value")
for label, value in label_mapping.items():
    print(label, ":", value)
```

```python
df.columns
```

```python
X = df[['itching', 'skin_rash', 'nodal_skin_eruptions', 'stomach_pain', 'skin_peeling', 'shivering', 'continuous_sneezing', 'lethargy']]
Y = df[['Ayurvedic Category', 'prognosis']]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
classifier = MultiOutputClassifier(RandomForestClassifier(n_estimators=100, random_state=42))
```

```python
classifier.fit(X_train, Y_train)
```

```python
Y_pred = classifier.predict(X_test)
```

```python
print(Y_pred)
```

```python
from sklearn.neural_network import MLPClassifier

# Initialize the Multi-Layer Perceptron classifier
mlp_classifier = MLPClassifier(hidden_layer_sizes=(100, 50), max_iter=1000, random_state=42)
```

```python
multioutput_classifier_mlp = MultiOutputClassifier(mlp_classifier)
```

```python
multioutput_classifier_mlp.fit(X_train, Y_train)
```

```python
Y_pred_mlp = multioutput_classifier_mlp.predict(X_test)
```

```python
from sklearn.svm import SVC
svm_classifier = SVC(kernel='linear', random_state=42)
```

```python
multioutput_classifier_svm = MultiOutputClassifier(svm_classifier)
```

```python
multioutput_classifier_svm.fit(X_train, Y_train)
```

```python
Y_pred_svm = multioutput_classifier_svm.predict(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_classifier = KNeighborsClassifier(n_neighbors=5)

# Initialize the multi-output classifier with k-NN as the base estimator
multioutput_classifier_knn = MultiOutputClassifier(knn_classifier)
```

```python
multioutput_classifier_knn.fit(X_train, Y_train)
```

```python
Y_pred_knn = multioutput_classifier_knn.predict(X_test)
```

```python
# Flatten the multi-output format
Y_test_kflat = Y_test.to_numpy().flatten()
Y_pred_kflat = Y_pred_knn.flatten()

# Print classification report
print("Classification Report:")
print(classification_report(Y_test_kflat, Y_pred_kflat))
```
